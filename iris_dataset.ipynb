```python
import pandas as pd
#loading the iris dataset
df=pd.read_csv("IRIS.csv")
#first 10 rows
print(df.head(10))
```

```python
#shape and datatypes
print(df.shape)
print(df.dtypes)
```

```python
print(df.describe())
```

```python
filtered_df = df.query(
    '`petal_length` > 4.5 and species == "virginica"'
)

print(filtered_df)
```

```python
grouped_stats = df.groupby('species').agg(
    avg_sepal_length=('sepal_length', 'mean'),
    max_petal_width=('petal_width', 'max'),
    std_sepal_width=('sepal_width', 'std')
)

print(grouped_stats)
```

```python
df['petal_ratio'] = df['petal_length'] / df['petal_width']

avg_petal_ratio = df.groupby('species')['petal_ratio'].mean()
print(avg_petal_ratio)
```
