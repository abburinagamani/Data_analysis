```python
import pandas as pd
df=pd.read_csv("raw_data.csv")

print(df.isnull())
print(df.isnull().sum())
```

```python
df2=df.copy()
df2["tax"]=df["income"].apply(lambda x:"20%" if x>=50000 else "10%")
df2
```

```python
#map 
gender_map={"Male":"M","Female":"F","Unknown":"U"}
df2["gender"]=df2["gender"].map(gender_map)
print(df2)
```

```python
#assign
df2=df2.assign(new_income=df["income"] *1.1)
print(df2)
```

```python
#replace
df2["country"]=df2["country"].replace("USA","US")
print(df2)
```

```python
#task we need to shift id column to end
df2=df.copy()
new_col_order=[col for col in df.columns if col!="id"]+["id"]
print(df[new_col_order])
```

```python
sorted_data=df2.to_csv("sorted_data.csv")
print(sorted_data)
```

```python
#grouping data
df.groupby("country")["income"].mean()
df.groupby("country")["income"].max()
df.groupby("country")["income"].min()
```

```python
df_customers=pd.DataFrame({
    "customer_id":[1,2,3,4],
    "name":["Adam","Bob","Charlie","Dave"]
})
df_orders=pd.DataFrame({
    "order_id":[101,102,103,104],
    "customer_id":[2,1,4,5],
    "amount":[250,120,300,180]
        
})
```

```python
df_customers
```

```python
pd.merge(df_customers,df_orders,on="customer_id")
```

```python
pd.merge(df_customers,df_orders,on="customer_id",how="left")
```

```python
pd.merge(df_customers,df_orders,on="customer_id",how="right")
```

```python
pd.merge(df_customers,df_orders,on="customer_id",how="outer")
```
